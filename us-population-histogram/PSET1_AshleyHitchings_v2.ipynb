```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
census = pd.read_csv("census1900-2000.csv")
```

```python
census_f = census[census["Sex"] == 2]
census_slice_f = census_f[["Year","Age", "People"]]
census_1900_f = census_slice_f[census_slice_f["Year"] == 1900].drop(columns=["Year"])
census_2000_f = census_slice_f[census_slice_f["Year"] == 2000].drop(columns=["Year"])

census_m = census[census["Sex"] == 1]
census_slice_m = census_m[["Year","Age", "People"]]
census_1900_m = census_slice_m[census_slice_m["Year"] == 1900].drop(columns=["Year"])
census_2000_m = census_slice_m[census_slice_m["Year"] == 2000].drop(columns=["Year"])
```

```python
total_pop_1900 = census[census["Year"] == 1900]["People"].sum()
total_pop_2000 = census[census["Year"] == 2000]["People"].sum()
```

```python
#proportion of total pop for the given year

plt.figure(figsize=(12,6))
plt.bar(census_2000_f["Age"] + 1, census_2000_f["People"]/total_pop_2000*100, color="purple", alpha = 0.6, label = "2000 - F")
plt.bar(census_1900_f["Age"], census_1900_f["People"]/total_pop_1900*100, color="pink", label = "1900 - F")

plt.xticks(np.arange(0, 95, 5))
plt.legend()
plt.xlabel("Age Group")
plt.ylabel("Percent of Population (%)")
plt.title("Proportional Age Distribution of U.S. Females in 1900 vs. 2000")

plt.figure(figsize=(12,6))
plt.bar(census_2000_m["Age"] + 1, census_2000_m["People"]/total_pop_2000*100, color="blue", alpha=0.5, label = "2000 - M")
plt.bar(census_1900_m["Age"], census_1900_m["People"]/total_pop_1900*100, color="skyblue", label = "1900 - M")
plt.xticks(np.arange(0, 95, 5))
plt.legend()
plt.xlabel("Age Group")
plt.ylabel("Percent of Population (%)")
plt.title("Proportional Age Distribution of U.S. Males in 1900 vs. 2000")
```
