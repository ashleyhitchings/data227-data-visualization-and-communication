```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
plt.style.use('fivethirtyeight')
import altair as alt
```

```python
dat = pd.read_csv("DEA_drugoverdose_data.csv")
```

```python
net_efficacy = dat["Number of Arrests"] - dat["Total Overdose Deaths"]
dat["Net Efficacy"] = net_efficacy
```

```python
# Devious graph

base = alt.Chart(dat).encode(
    alt.X('Fiscal Year:O', axis=alt.Axis(title="Fiscal Year"))
).properties(
    width=500,
    height=300,
)

area = base.mark_area(opacity=0.3, color='red').encode(
    alt.Y('Net Efficacy',
        axis = alt.Axis(title='Net Efficacy', titleColor='maroon', grid=True))
)


line = base.mark_line(stroke='maroon', interpolate='monotone').encode(
    alt.Y('Net Efficacy',
         axis=alt.Axis(title=None, ticks=False, labels=False))
)
    
line2 = base.mark_line(stroke='green', interpolate='monotone').encode(
    alt.Y('Budget ($B)',
          axis=alt.Axis(title='DEA Annual Budget ($B)', titleColor='green'),
          scale=alt.Scale(domain=(1.4, 3.2)))
)

alt.layer(area, line, line2).resolve_scale(
    y = 'independent'
).properties(
    title='Drug Enforcement Administration Efficacy Plumments With Soaring Expenditures',
    width=500,
    height=300,
)
```

```python
# base encoding the x-axis
base = alt.Chart(dat).encode(
    alt.X('Fiscal Year:O', axis=alt.Axis(title="Fiscal Year", grid=True))
).properties(
    width=500,
    height=450,
    title="Drug Enforcement Administration Expenditures vs. Arrests and Drug-Related Deaths"
)


arrests = base.mark_line(color='#cc393e', strokeWidth=2.5).encode(alt.Y("Number of Arrests", axis = alt.Axis(title="Number of Occurrences", grid=True)))
pre_opioids = base.mark_line(color='#a3abe4', strokeWidth=2.5).encode(y="Prescription Opioids")
heroin = base.mark_line(color='#8dc4c0', strokeWidth=2.5).encode(y="Heroin")
cocaine = base.mark_line(color='#4b78a8', strokeWidth=2.5).encode(y="Cocaine")
meth = base.mark_line(color='#eeca3b', strokeWidth=2.5).encode(y="Methamphetamine")
benz = base.mark_line(color='#f58518', strokeWidth=2.5).encode(y="Benzodiazepines")
antidep = base.mark_line(color='#e274ce', strokeWidth=2.5).encode(y="Antidepressants")

left_data = (arrests + pre_opioids + heroin + cocaine + meth + benz + antidep).resolve_scale(y="shared")

right_data = base.mark_line(color='green', opacity=0.5, strokeWidth=5).encode(alt.Y("Budget ($B)", title="DEA Annual Budget ($B)",
                                                       axis=alt.Axis(title='DEA Annual Budget ($B)', titleColor='black')))

alt.layer(left_data, right_data).resolve_scale(y="independent")
```
